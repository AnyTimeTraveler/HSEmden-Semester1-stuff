```python
import numpy as np
import scipy as sci
import seaborn as sea
import matplotlib.pyplot as plt
import pandas as pd
```

Importieren des Beispieldatensatzes mittels pandas

```python
df = pd.read_csv('/home/jme/PycharmProjects/sci_kit_learn_test/Iris_data_set.txt', sep=',', index_col=False)
```

### Anlegen der HS EL Farbpalette

```python

hs_el_colors = ['#016AAC',  # Light Blue
                '#013F68',  # Dark Blue
                '#A400AF',  # Light Purple
                '#63006A',  # Dark Purple
                '#D3F800',  # Light green
                '#869D00',  # Dark green
                '#FF8E00',  # Light gray
                '#A25A00',  # Dark gray
                ]
id3as_colors = ['#e85511',  # orange
                '#005a9f',  # blue
                '#63b338',  # green
                '#8e340b',  # Dark orange
                '#0091ff',  # Dark blue
                '#417425',  # Dark green
                ]
```

### Einfacher XY Scatter Plot mit Seaborn

* x='sepal_length' : Feature für die X-Achse aus df

* y='petal_length' : Feature für die y-Achse aus df

* data=df : Verweis auf das zuvor geladene Datenframe

* fit_reg=False : Keine lineare Fitting Linie darstellen

* hue='ground_truth' : Verweis auf die Zeile in der die Labels für den Datensatz stehen 

* palette=hs_el_colors : Verwenden der zuvor angelegten HS EL Farbpalette

```python
sea.lmplot(x='sepal_length', y='petal_length', data=df, fit_reg=False,hue='ground_truth', palette=hs_el_colors)
```

### Violinen Plot

![Violine Plot](img/violin_plot.png)

https://datavizcatalogue.com/methods/violin_plot.html


```python
# set whitegrid theme
sea.set_style('whitegrid')

# Violin plot
sea.violinplot(data=df,palette=hs_el_colors)

#Customize with Matplotlib
plt.title('Violine Plot')

plt.xlabel('Häufigkeit')

plt.ylabel('Feature')

plt.show()
```

### Box & Whisker Plot

![Box Plot](img/box_plot.png)

https://datavizcatalogue.com/methods/box_plot.html 

```python
sea.boxplot(data=df,palette=hs_el_colors)
```

### Swarm Plot

```python
sea.swarmplot(x='ground_truth', y='petal_width', data=df, palette=hs_el_colors)
```

### Histogramm 

```python
# Distribution Plot (a.k.a. Histogram)
sea.distplot(df.sepal_length)
```

### Bar Plot

```python
sea.countplot(x='ground_truth', data=df, palette=hs_el_colors)
# Rotate x-labels
plt.xticks(rotation=-45)
```

### Pair Plot
Darstellung der Beziehungen zwischen Features

```python
iris = sea.load_dataset("iris")
sea.pairplot(iris);
```
